# Punto 4 Construir una tabla que muestre cuántas medallas obtuvieron los hombres en total en cada año que se realizó el evento.

```python
df2.head() #Vemos los nombres de las columnas y obtenemos que su nombre es Event_gender
```

```python
dfHombres = df2[df2.Event_gender == 'M'] #Creamos un dataframe con solo los hombres
dfHombres
```

```python
dfTotalHombres = dfHombres.groupby(['Year', 'Event_gender'])['Event_gender'].count() #Monstramos el dataframe con solo el Año y el Género
dfTotalHombres.to_frame()
```
